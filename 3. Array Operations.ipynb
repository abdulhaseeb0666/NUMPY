**Broadcasting**

```python
lst = [10,20,30,40]
arr = np.array(lst)
```

```python
for i in lst:
    print(i + 10)
```

```python
arr + 10
```

**Deep and Shallow Copy**

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice
```

```python
slice * 10
```

```python
a  #Changes in slice does not effect the real array And it is a Shallow Copy
```

```python
b = a 
```

```python
b[0] = 99
```

```python
a  #Changes in b affected the real array And it is a Deep Copy
```

```python
c = a.copy()
c[0] = 1000
c
a  #Changes in c does not effect the real array And it is a Shallow Copy
```

```python
mat1 = np.array([[1,2] , [3,4]])
mat2 = np.array([[5,6] , [7,8]])
```
