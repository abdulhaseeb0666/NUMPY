```python
import numpy as np
```

**Stacking Array**

```python
arr1 = np.arange(1,6)
arr2 = np.arange(6,11)
```

```python
np.vstack((arr1 , arr2))
```

```python
np.hstack((arr1 , arr2))
```

```python
np.column_stack((arr1 , arr2))
```

**Splitting Arrays**

```python
arr3 = np.arange(16).reshape(4,4)
arr3
```

```python
np.hsplit(arr3 , 2)
```

```python
np.hsplit(arr3 , 4)
```

```python
np.vsplit(arr3 , 2)
```

```python
np.vsplit(arr3 , 4)
```

```python
a = np.vsplit(arr3 , 4)
```

```python
for i in a:
    print(i)
```
