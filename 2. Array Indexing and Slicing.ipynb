```python
import numpy as np
```

**Numpy Indexing and Slicing of Vectors**

```python
arr = np.arange(11 , 21)
arr
```

```python
arr[1 : 5] # It is just like slicing lists in Python
```

```python
arr[1:]
```

```python
arr[1: :2]  #The 2 after the second column shows that it will take steps of two from value at 1st index to the last item in the array
```

```python
arr #The above commands will not effect the original array
```

**Numpy Indexing and Siicing of Matix**

```python
arr2 = np.arange(1 , 31).reshape(6 , 5)
arr2
```

```python
arr2[0]
```

```python
arr2[1]
```

```python
arr2[0,2]  #It will give the element at 2nd index from the array at 0 index
```

```python
arr2[4,3] #It will give the element at 3rd index from the array at 4th index
```

```python
arr2
```

```python
#The value before coma represents the rows and the value after coma represents the columns
# It means, in the following example we are slicing 3rd,4th rows and 2nd,3rd columns
arr2[3:5 , 2:4]
```

```python
arr2[3:6 , 3:5]
```

```python
arr2[3: , 3: ]
```
