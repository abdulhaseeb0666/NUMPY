```python
import numpy as np
```

Basic Array Creation

```python
arr = np.array([1,2,3,4])
arr
```

```python
lst = [2, 3.5]
arr2 = np.array(lst)
arr2  #integer is converted to float because arrays can have only one data type
```

```python
lst2 = [2 , 3.5 , "string"]
arr3 = np.array(lst2)
arr3  #Here integera and float both are converted to string
```

```python
lst2 = [2 , 3.5 , True , False]
arr3 = np.array(lst2)
arr3  #Here integer and Boolean both are converted to Float
```

```python
lst4 = [[1,2,3] , [4,5,6] , [7,8,9]]
# It is a List
print(lst4)
multiDArray = np.array(lst4)
multiDArray 
```

Array Generation Function

```python
arr5 = np.arange(1, 11 , 2)
arr5
```

```python
arr6 = np.zeros(6) #It will creat a 1-D Array
print(arr6)
print("\n")

arr7 = np.zeros((6,6))  #It will create an array of order 6*6 containing only zeros
print(arr7)
```

```python
arr8 = np.linspace(1,5,3)  
print(arr8)
arr9 = np.linspace(1,5,5)
print(arr9)
arr10 = np.linspace(0 , 1, 10)
print(arr10)
```

Random Generation Function

```python
np.random.rand(10)  #It will create 10 elements between 0 and 1
```

```python
np.random.randn(10) #It will create 10 elements between -3 and 3
```

```python
np.random.randint(10) #It will create 1 element between 0 and 10
```

```python
np.random.randint(10 , 20 , 5) #It will create 5 elements between 10 and 20
```
