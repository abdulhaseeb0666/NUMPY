**Valid Sudoku**

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
# Sum of all columns must be 45
b = np.sum(s , axis = 1)
cond = True
for i in b:
    if(i != 45) :
        print("Suduko is not Valid...\n")
        cond = False
        break
if(cond == True):
    print("Suduko is Valid")
        
```

```python
# Sum of all rows must be 45
c = np.sum(s , axis = 0)
cond = True
for i in c:
    if(i != 45) :
        print("Suduko is not Valid...\n")
        cond = False
        break
if(cond == True):
    print("Suduko is Valid")
```

```python
## Sum of all 3*3 blocks must be 45
for i in range(0,9,3):
    for j in range(0,9,3):
        n = s[i:i+3 , i:i+3]
        print(n.sum())
```

**General Questions**

```python
# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
np.shape(data)
```

```python
#Find the average age of students.
np.mean(data , axis = 0)[0]
```

```python
#Extract Math marks of all students.
data[ : , 1:2]
```

```python
#Find the highest Science mark.
np.max(data , axis = 1)
```

```python
#Get details of the student who scored more than 90 in Math.
for i in data[ : , 1:2]:
    for j in i:
        if(j>90):
            print(j)
```

```python
#Increase Math marks of all students by 5.
data[ : , 1:2] + 5
```

```python
#Find how many students are younger than 19.
count = 0
for i in data[ : , 0:1]:
    for j in i:
        if(j>19):
            count += 1
print(count)
```

```python
#Calculate the average marks in each subject (column-wise mean).
marks = np.mean(data , axis=0)
print(f"Average Marks for Maths: {marks[1]}")
print(f"Average Marks for Maths: {marks[2]}")
```

```python
#Get data of students who scored at least 80 in both subjects.
for i in data[ : , 1:]:
    condition = True
    for j in i:
        if(j<80):
            condition = False
    if(condition == True):
        print(i)
```
